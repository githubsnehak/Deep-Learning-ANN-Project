```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
data= pd.read_csv("Churn_Modelling.csv")
data.head(245)
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(x,y)
```

```python
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xTrain.shape
```

```python
xTest
```

```python
import tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout
from keras.callbacks import EarlyStopping
from keras.regularizers import l2
```

```python
ann=Sequential()
```

```python
ann.add(Dense(6,input_dim=xTrain.shape[1],activation='relu',kernel_regularizer=l2(0.01)))
#ann.add(BatchNormalization()) # do normalization to hidden layers (do scaling to hidden layers to avoid gradient decent prob)
#ann.add(Dropout(0.5)) #  50% random data wil not participate and rest 50%will be given as input
ann.add(Dense(4,activation='relu'))
#ann.add(BatchNormalization())
#ann.add(Dropout(0.5))
ann.add(Dense(2,activation='relu'))
#ann.add(BatchNormalization())
#ann.add(Dropout(0.5))
ann.add(Dense(1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=ann.fit(xTrain,yTrain, batch_size=100,epochs=50,validation_data=(xTest,yTest),callbacks=EarlyStopping())  # jaha bhi accuracy mein difference aayga vaha earystopping rok dega
history
```

```python
train_accuracy=history.history['accuracy']
test_accuracy=history.history['val_accuracy']
 
```

```python
# Predict the labels for the test data
yPred = ann.predict(xTest)

# Convert predictions to binary values (0 or 1)
yPred_binary = (yPred > 0.5).astype(int)

# Calculate metrics

# Accuracy
accuracy = accuracy_score(yTest, yPred_binary)
accuracy
```

```python
#Precision Score
PrecisionScore=precision_score(yTest, yPred_binary)
PrecisionScore
```

```python
#Recall Score
RecallScore=recall_score(yTest, yPred_binary)
RecallScore
```

```python
#F1 Score
F1Score=f1_score(yTest, yPred_binary)
F1Score
```
